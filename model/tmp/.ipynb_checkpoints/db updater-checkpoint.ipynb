```python
def rebuild_price(prices, field):
    d = {}
    for asset in prices.index.levels[1]:
        p_ = prices[[field,'underlying']].xs(asset, level=1)
        r_ = p_.pct_change()
        p_rebuild = (r_[field].combine_first(r_.underlying).fillna(0)+1).cumprod()
        d[asset] = p_rebuild / p_rebuild[-1] * p_[field][-1]
        
    return pd.DataFrame(d)
```
